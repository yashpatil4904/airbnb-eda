```python
from sklearn.impute import KNNImputer
import pandas as pd
df=pd.read_csv('/content/west_coast dataset.csv')
df.shape


df.sample(5)
```

checking for duplicates

```python
print(len(df)-len(df.drop_duplicates()))
#dropping unneccessary column
columns_to_drop = ['calendar_updated','listing_url','scrape_id','source','picture_url','host_id','host_url','host_thumbnail_url','host_picture_url','neighbourhood']
df = df.drop(columns=columns_to_drop, errors='ignore')
df.shape
df.info()
```

```python
#calculate the percentage of missing value
round(df.isnull().sum()/len(df)*100,2).sort_values(ascending=False)
```

From this I can see that license,neighbourhood_group_cleansed,neighbourhood_overview,host_about

so dropping those column as well

```python
column_with_large_missing_values=['license','neighbourhood_group_cleansed','neighborhood_overview','host_about']
df=df.drop(columns=column_with_large_missing_values,errors='ignore')
df.info()
```

converting features into there preferable d types

```python

def is_datelike(s, fmt='%Y-%m-%d'):
    try:
        pd.to_datetime(s, format=fmt, errors='raise')
        return True
    except ValueError:
        return False
df['last_scraped'] = pd.to_datetime(df['last_scraped'].apply(lambda x: x if is_datelike(x) else pd.NaT), errors='coerce')
df['host_since'] = pd.to_datetime(df['host_since'], format='%Y-%m-%d', errors='coerce')
df['last_review'] = pd.to_datetime(df['last_review'], format='%Y-%m-%d', errors='coerce')
df.head()
```

```python
# Columns to convert
percentage_columns = ['host_response_rate', 'host_acceptance_rate']

for column in percentage_columns:
    try:
        df[column] = df[column].str.rstrip('%').astype(float) / 100
        df[column] = df[column].astype(float)
    except (ValueError, AttributeError):
        print(f"Could not convert column '{column}' to percentage. Skipping...")
# Convert INTO BOOLEAN
for column in ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'has_availability', 'instant_bookable']:
    df[column] = df[column].map({'t': True, 'f': False}).astype(bool)

# Convert numeric columns, handling potential errors
for column in ['host_listings_count', 'host_total_listings_count', 'minimum_nights', 'maximum_maximum_nights', 'minimum_maximum_nights']:
    df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64')

# Convert 'price' to float after removing currency symbols
df['price'] = df['price'].astype(str).str.replace('[\$,]', '', regex=True).astype(float)
df.rename(columns={'price': 'price_in_dollar'}, inplace=True)

# Print the updated DataFrame info to check data types
df.info()
df.head()
```

General datacleaning step handling mising values

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['review_scores_checkin'] = df['review_scores_checkin'].fillna(df['review_scores_checkin'].median())
df['review_scores_value'] = df['review_scores_value'].fillna(df['review_scores_value'].median())
df['review_scores_location'] = df['review_scores_value'].fillna(df['review_scores_value'].median())
df['review_scores_communication'] = df['review_scores_value'].fillna(df['review_scores_value'].median())
df['review_scores_cleanliness'] = df['review_scores_value'].fillna(df['review_scores_value'].median())
df['review_scores_accuracy'] = df['review_scores_value'].fillna(df['review_scores_value'].median())
df['review_scores_rating'] = pd.to_numeric(df['review_scores_rating'], errors='coerce')

imputer = KNNImputer(n_neighbors=5)  # Adjust n_neighbors as needed
df['review_scores_rating'] = imputer.fit_transform(df[['review_scores_rating']])[:, 0]
# Replace missing values in 'first_review' with 'host_since'
df['first_review'] = df['first_review'].fillna(df['host_since'])
# Replace missing values in 'last_review' with 'last_scraped'
df['last_review'] = df['last_review'].fillna(df['last_scraped'])
#replacing host location with city
df['host_location'] = df['host_location'].fillna(df['city'])
df['host_response_time'] = df['host_response_time'].fillna(df['host_response_time'].mode()[0])
df['host_response_rate'] = df['host_response_rate'].fillna(df['host_response_rate'].mode()[0])
df['host_neighbourhood'] = df['host_neighbourhood'].fillna('no neighbourhood specified')
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna(df['host_acceptance_rate'].median())
df['beds'] = df['beds'].fillna(df['beds'].mode()[0])
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].mode()[0])
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mode()[0])
df['price_in_dollar'] = df['price_in_dollar'].fillna(df['price_in_dollar'].mean())
df['description'] = df['description'].fillna('no Description')
# print(df[df['bathrooms_text'].isnull()])
#after analysis we can observe that row of index 2434 and 4549 have maximum number of missing value
index_to_drop = [2434, 4549]
existing_indices = df.index
index_to_drop = [index for index in index_to_drop if index in existing_indices]
if index_to_drop:
    df.drop(index_to_drop, inplace=True)
df['bathrooms_text'] = df['bathrooms_text'].fillna('unknown')
#droping row where city is null
df = df.dropna(subset=['city'])
round(df.isnull().sum()/len(df)*100,2).sort_values(ascending=False)
```

```python
# Specify the file path and name
file_path = '/content/sample_data/missingvalues_handeled.csv'  # Changed to a file name within the directory

#Convert DataFrame to CSV
df.to_csv(file_path, index=False)
```
